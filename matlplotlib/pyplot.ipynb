# Matplotlib


Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. 

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Basics

```python
# Sample data
x = [1, 2, 3]
```

### Single Line

```python
plt.plot(x)
plt.show()
```

### Multiple Line

```python
y = [1, 4, 9]
```

You can plot as many lines as you like by simply adding more plt.plot() functions:

```python
plt.plot(x)
plt.plot(y)
plt.show()
```

### X and Y Line

You can also plot many lines by adding the points for the x- and y-axis for each line in the same plt.plot() function.

```python
# In a single .plot function
plt.plot(x,y)
plt.show()
```

### Labels and Title

With Pyplot, you can use the **xlabel()** and **ylabel()** functions to set a label for the x- and y-axis.

The `title()` function add a title for the plot

the `xlabel()` function adds a label along the horizontal x-axis

the `ylabel()` function adds a label along the vertical y-axis

the `legend()` function adds a legend for each line

the `plot()` function accepts variables that will be appear in the plot

```python
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend("y")
plt.show()
```

### Markers

 You can use the keyword argument **marker** to emphasize each point with a specified marker

```python
plt.plot(x,y, marker = 'o')
plt.show()
```

### Linestyles

You can use the keyword argument **linestyle**, or shorter **ls**, to change the style of the plotted line:

```python
plt.plot(x,y, linestyle = 'dotted')
plt.show()
```

### Grid

With Pyplot, you can use the `grid()` function to add grid lines to the plot.

```python
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend("y")
plt.grid()
plt.show()
```

You can use the keyword argument **axis** to only apply a grid to tbe x or y axis

```python
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend("y")
plt.grid(axis='x')
plt.show()
```

```python
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend("y")
plt.grid(axis='y')
plt.show()
```

You can use the keyword argument **color**, **linestyle** and **linewidth** to customize the grid

```python
plt.plot(x, y)
plt.title("test plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend("y")
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()
```

## Plot a real world dataset

```python
data = pd.read_csv('countries.csv')
```

```python
data
```

### Compare the population growth in the US and China

```python
us = data[data.country == 'United States']
```

```python
us
```

```python
china = data[data.country == 'China']
```

```python
china
```

### Plot the population growth between US and China

```python
plt.plot(us.year, us.population)
plt.plot(china.year, china.population)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

Divide the population by 1 million

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
us.population
```

Divide the population each year by 100

```python
us.population / us.population.iloc[0] * 100
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population growth (first year = 100)')
plt.show()
```

# Scatterplot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('countries.csv')
```

```python
china = data[data.country == 'China']
```

```python
china
```

```python
chinax = china.year
chinay = china.population

plt.scatter(chinax, chinay)
```

```python
us = data[data.country == 'United States']
```

```python
usx = us.year
usy = us.population

plt.scatter(usx, usy)
```

```python
chinax = china.year
chinay = china.population

plt.scatter(chinax, chinay, color='red')

usx = us.year
usy = us.population

plt.scatter(usx, usy, color='blue')
plt.show()
```
